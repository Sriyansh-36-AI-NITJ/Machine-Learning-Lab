```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, centers=4, cluster_std=0.60, random_state=0)
df = pd.DataFrame(X, columns=['cgpa', 'iq'])
X = df.iloc[:,:].values

plt.scatter(X[:,0], X[:,1])
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.show()
```

```python
wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
print(y_means)
```

```python
# Cluster 0 (Blue)
plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], color='blue', label='Cluster 0')

# Cluster 1 (Red)
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], color='red', label='Cluster 1')

# Cluster 2 (Green)
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], color='green', label='Cluster 2')

# Cluster 3 (Yellow)
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], color='yellow', label='Cluster 3')

plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.legend()
plt.show()
```

```python
# Generating 3D data
from sklearn.datasets import make_blobs
import plotly.express as px

X_3d, _ = make_blobs(n_samples=200, centers=4, n_features=3, cluster_std=1.5, random_state=42)
df_3d = pd.DataFrame(X_3d, columns=['Feature 1', 'Feature 2', 'Feature 3'])

km_3d = KMeans(n_clusters=4)
y_means_3d = km_3d.fit_predict(X_3d)

df_3d['Cluster'] = y_means_3d

fig = px.scatter_3d(df_3d, x='Feature 1', y='Feature 2', z='Feature 3', color='Cluster')
fig.show()
```
