```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08

    print(f"{'m':<15} {'b':<15} {'cost':<15} {'iteration':<10}")
    print("-" * 60)

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr

        # Calculate Cost (Mean Squared Error)
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])

        # Calculate Partial Derivatives (Gradients)
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)

        # Update Steps (adjusting m and b)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        # Print progress every 1000 iterations to avoid flooding the output
        if i % 1000 == 0:
             print(f"{m_curr:<15.6f} {b_curr:<15.6f} {cost:<15.6f} {i:<10}")

    return m_curr, b_curr
```

```python
m_final, b_final = gradient_descent(x, y)
print(f"\nFinal Result -> m: {m_final:.4f}, b: {b_final:.4f}")
```

```python
import matplotlib.pyplot as plt

# Plot actual data points
plt.scatter(x, y, color='red', marker='+', s=100, label='Actual Data')

# Plot the learned line
y_predicted_line = m_final * x + b_final
plt.plot(x, y_predicted_line, color='blue', label='Prediction Line')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
