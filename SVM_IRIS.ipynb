```python
from google.colab import files
import pandas as pd
uploaded = files.upload()
data = pd.read_csv('iris.csv')
print(data.head())
```

```python
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]
```

```python
from sklearn import svm
model = svm.SVC()
model.fit(X.values, Y)
print(model.score(X.values, Y))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X1 = X_pca[:, 0]
X2 = X_pca[:, 1]
```

```python
P = []
E = []
L = 0

for i in range(len(X)):
    row = X.values[i]
    prediction = model.predict([row])
    P.append(prediction[0])

    if prediction[0] != Y[i]:
        E.append(1)
        L = L + 1
    else:
        E.append(0)

print(L)
```

```python
import matplotlib.pyplot as plt
color_mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
P_numeric = [color_mapping[label] for label in P]
plt.scatter(X1, X2, c=P_numeric, cmap='viridis')
plt.show()
```
